## Valeurs manquantes

```python
missing_values = database.isnull().sum()
print(missing_values.sum())
```

```python
missing_percent = missing_values/len(database)*100
```

```python
missing_percent.sort_values(0, False)
```

Il y a de nombreuses colonnes qui ne sont pas remplies ou peu remplies.
